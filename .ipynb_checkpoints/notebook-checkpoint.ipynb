```python
num_of_records = len(df)
num_of_fraud = len(df.query('Class == 1'))
print('Dataset contains {} records'.format(num_of_records))
print('Of those, there are {} fraudulent transactions'.format(num_of_fraud))
print('\n{0:1.3f}% of transactions are fraudulent'.format(num_of_fraud/num_of_records * 100))
```

# Note: Very imbalanced dataset

# Exploratory Data Analysis

```python
fig, ax = plt.subplots(figsize=(25,10))

# Check correlations
corr = df.corr()
sns.heatmap(corr, cmap='bwr', annot = True, fmt = '1.3f', vmin=-1, vmax=1, ax=ax)
ax.set_title("Correlations", fontsize=14)
plt.show()
```

# Interesting, from the correlation matrix it seems that the transaction amount is not correlated with Class.

```python
df_corr = df.corr()
class_corr = df_corr.iloc[-1:, :]

# col_names = class_corr.columns.to_numpy()
# col_values = class_corr.to_numpy()

class_corr
```

```python
fig, ax = plt.subplots(figsize=(20,25))

nrows = 8
ncols = 4

df_fake = df.query('Class == 1')

for i, feature in enumerate(df.columns):
    ax = fig.add_subplot(nrows,ncols,i+1)
    df[feature].hist(bins = 20, ax = ax, facecolor= 'blue', label = 'All Transactions')
    df_fake[feature].hist(bins = 20, ax = ax, facecolor= 'red', label = 'Fraudulent Transactions')
    ax.set_title(feature + " Distribution", color = 'k')
    ax.set_yscale('log')
    ax.legend()
    
fig.tight_layout()  
plt.show()
```
